```python
## import pandas
import pandas as pd
```

```python
## Display floating numbers with 2 decimal places
pd.options.display.float_format = '{:.2f}'.format
```

```python
## read the data
pd.read_excel("data_raw/HHS_Grants_Terminated_April_2025_Mechanism_Isolated.xlsx", sheet_name = "Sheet1")
```

```python
## Store the dataset in a variable
df_NIH_Terminated_Grants_Family = pd.read_excel("data_raw/HHS_Grants_Terminated_April_2025_Mechanism_Isolated.xlsx", sheet_name = "Sheet1")
```

```python
## Call the dataset
df_NIH_Terminated_Grants_Family
```

```python
## Create a pivot table that counts the number of terminated grants in each grant family
## Store the pivot table in a variable
NIH_Terminated_Grants_Family_Pivot_Table = df_NIH_Terminated_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Award Number'], aggfunc = 'count')
```

```python
## Sort the pivot table
NIH_Terminated_Grants_Family_Pivot_Table.sort_values(by='Award Number', ascending=False)
```

```python
## Create a pivot table that both counts the number of grants in each grant family and sums the total amount obligated for each grant family
## Store the pivot table in a variable
NIH_Terminated_Grants_Family_Pivot_Table_Funding = df_NIH_Terminated_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Award Number', 'Total Amount Obligated'], aggfunc = {'Award Number': 'count',
                                                                                                                            'Total Amount Obligated': 'sum',
                                                                                                                            })
```

```python
## Call the pivot table
NIH_Terminated_Grants_Family_Pivot_Table_Funding
```

```python
## Look at data of NIH active grants broken down by family
pd.read_excel("data_raw/NIH_Reporter_Active_Grants_Summary_Family_Isolated_4.8.25.xlsx")
```

```python
## Store the dataset in a variable
df_NIH_Active_Grants_Family = pd.read_excel("data_raw/NIH_Reporter_Active_Grants_Summary_Family_Isolated_4.8.25.xlsx")
```

```python
## call the variable
df_NIH_Active_Grants_Family
```

```python
## Create a pivot table that counts the number of grants in each grant family
## Store the pivot table in a variable
df_NIH_Active_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Projects'], aggfunc = 'sum')
NIH_Active_Grants_Family_Pivot_Table = df_NIH_Active_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Projects'], aggfunc = 'sum')
```

```python
## Call the pivot table
NIH_Active_Grants_Family_Pivot_Table
```

```python
## Create a pivot table that counts the number of grants in each grant family and sums the total funding for each grant family
## Store the pivot table in a variable
df_NIH_Active_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Projects', 'Total Funding Including Sub Projects'], aggfunc = {'Projects': 'sum',
                                                                                                                                     'Total Funding Including Sub Projects': 'sum'})
NIH_Active_Grants_Family_Funding_Pivot_Table = df_NIH_Active_Grants_Family.pivot_table(index = ['Grant Family'], values = ['Projects', 'Total Funding Including Sub Projects'], aggfunc = {'Projects': 'sum',
                                                                                                                                     'Total Funding Including Sub Projects': 'sum'})
```

```python
## Call the pivot table
NIH_Active_Grants_Family_Funding_Pivot_Table
```

```python
## Export active grants funding pivot table to csv
NIH_Active_Grants_Family_Funding_Pivot_Table.to_csv('active_grants_funding.csv')
```

```python
## Export terminated grant family pivot table to csv
NIH_Terminated_Grants_Family_Pivot_Table_Funding.to_csv('terminated_grants_funding.csv')
```
