```python
## import pandas
import pandas as pd
```

```python
## Display floating numbers with 2 decimal places
pd.options.display.float_format = '{:.2f}'.format
```

```python
## read DOGE grant termination data
pd.read_excel("Raw_Data/DOGE_HHS_Grants_Terminated_4.30.25.xlsx")
```

```python
## Store the dataset in a variable
df_DOGE_Terminated_Grants = pd.read_excel("Raw_Data/DOGE_HHS_Grants_Terminated_4.30.25.xlsx")
```

```python
## get info on the dataset
df_DOGE_Terminated_Grants.info()
```

```python
## Filter the data to only NIH grants
df_DOGE_Terminated_Grants.query("Awarding_Office == 'NIH'")
```

```python
## Save filtered data in a variable
## Add .copy to avoid 'Setting with Copy' error that occurs when you are saving a slice of a dataframe to a variable 
## Call the filtered data to make sure it works
df_NIH_DOGE_Terminated_Grants = df_DOGE_Terminated_Grants.query("Awarding_Office == 'NIH'").copy()
df_NIH_DOGE_Terminated_Grants
```

```python
## Remove duplicate awards based on FAIN number
df_NIH_DOGE_Terminated_Grants.drop_duplicates(subset=['FAIN'])
```

```python
## Save clean data in a variable
## Add .copy to avoid 'Setting with Copy' error that occurs when you are saving a slice of a dataframe to a variable 
## Call the cleaned data to make sure it works
df_NIH_DOGE_Terminated_Clean = df_NIH_DOGE_Terminated_Grants.drop_duplicates(subset=['FAIN']).copy()
df_NIH_DOGE_Terminated_Clean
```

```python
## get info on the dataset to understand how many NIH grants have been terminated
df_NIH_DOGE_Terminated_Clean.info()
```

```python
## Create columns that isolates grant mechanism and grant family
## Function to extract the left n characters of a string
def left(text,n):
    return text[:n]
```

```python
## Create new columns using the left function
## Call the dataset to make sure it works
df_NIH_DOGE_Terminated_Clean['Grant_Mechanism'] = df_NIH_DOGE_Terminated_Clean['FAIN'].apply(lambda x: left(x,3))
df_NIH_DOGE_Terminated_Clean['Grant_Family'] = df_NIH_DOGE_Terminated_Clean['FAIN'].apply(lambda x: left(x,1))
df_NIH_DOGE_Terminated_Clean
```

```python
## Reorder columns
## Call columns into a list
list(df_NIH_DOGE_Terminated_Clean.columns)
```

```python
## reorder and call
df_NIH_DOGE_Terminated_Clean = df_NIH_DOGE_Terminated_Clean[[
'Awarding_Office',
'Grant_Mechanism',
'Grant_Family',
 'FAIN',
 'Award_Number',
 'Recipient_Name',
 'Date_Terminated',
 'Total_Amount_Obligated',
 'Total_Amount_Expended',
 'Total_Payment_Amount',
 'Unliquidated_Obligation',
 'Award_Title',
 'Presidential_Action',
 'For_Cause',]]
df_NIH_DOGE_Terminated_Clean
```

```python
## Determine the number of terminated grants by award family
df_NIH_DOGE_Terminated_Clean.groupby("Grant_Family")["Award_Number"].count().to_frame("Number of Grants")
```

```python
## Save as a new dataframe
df_DOGE_Clean_Grant_Family = df_NIH_DOGE_Terminated_Clean.groupby("Grant_Family")["Award_Number"].count().to_frame("Number of Grants")
```

```python
## read NIH active grants data
pd.read_excel("Raw_Data/NIH_Reporter_Activity_Code_Summary_5.1.25.xlsx")
```

```python
## Store the dataset in a variable
df_NIH_Reporter_Grants_Data = pd.read_excel("Raw_Data/NIH_Reporter_Activity_Code_Summary_5.1.25.xlsx")
```

```python
## get info on the dataset
df_NIH_Reporter_Grants_Data.info()
```

```python
## Create column that isolates grant family
## Function to extract the left n characters of a string
def left(text,n):
    return text[:n]
```

```python
## Create new column using the left function
## Call the dataframe
df_NIH_Reporter_Grants_Data['Grant_Family'] = df_NIH_Reporter_Grants_Data['Activity Code'].apply(lambda x: left(x,1))
df_NIH_Reporter_Grants_Data
```

```python
## Reorder columns
## Call columns into a list
list(df_NIH_Reporter_Grants_Data.columns)
```

```python
## reorder and call
df_NIH_Reporter_Grants_Data = df_NIH_Reporter_Grants_Data[[
'Grant_Family',
'Activity Code',
'Projects',
'Total Funding',
'Sub Projects',
'Sub Project Funding',
]]
df_NIH_Reporter_Grants_Data
```

```python
## Determine the number of active grants by award family
df_NIH_Reporter_Grants_Data.groupby("Grant_Family")["Projects"].sum().to_frame("Number of Grants")
```

```python
## Save as a new dataframe
df_NIH_Active_Grant_Family = df_NIH_Reporter_Grants_Data.groupby("Grant_Family")["Projects"].sum().to_frame("Number of Grants")
```

```python
## Merge datasets on grant family
df_DOGE_Clean_Grant_Family.merge(df_NIH_Active_Grant_Family, on = "Grant_Family")
```

```python
## Save as a new dataframe
df_Merged_Grant_Family  = df_DOGE_Clean_Grant_Family.merge(df_NIH_Active_Grant_Family, on = "Grant_Family")
```

```python
## Call the new dataframe
df_Merged_Grant_Family
```

```python
## to rename columns
df_Merged_Grant_Family.rename(columns = {"Number of Grants_x": "Terminated_Grants", "Number of Grants_y": "Active_Grants"}, inplace = True)
```

```python
## Recall the dataframe
df_Merged_Grant_Family
```

```python
## Create a new column that sums the terminated grants and active grants
## Call the updated dataframe
df_Merged_Grant_Family["Total_Grants"] = df_Merged_Grant_Family["Terminated_Grants"] + df_Merged_Grant_Family["Active_Grants"]
df_Merged_Grant_Family
```

```python
## Create a new column that calculates the percent termination for each grant family based on the volume of active grants pre-termination
## Call the updated dataframe
df_Merged_Grant_Family["Percent Terminated"] = (df_Merged_Grant_Family["Terminated_Grants"]/df_Merged_Grant_Family["Total_Grants"])* 100
df_Merged_Grant_Family
```

```python
## Sort the dataframe by total grants (ascending to descending) as you want to look at percent termination for grant families with the highest total number of grants
df_Merged_Grant_Family.sort_values("Total_Grants", ascending = False)
```

```python
## Add additional column relecting percent of grants active following termination for Datawrapper visulization purposes
## Call updated dataframe
df_Merged_Grant_Family["Percent Active Post Termination"] = 100 - (df_Merged_Grant_Family["Percent Terminated"])
df_Merged_Grant_Family
```

```python
df_Merged_Grant_Family.to_csv('NIH Terminated Grant Analysis Results.csv', index=True)
```
